```python
from pathlib import Path
```

```python
data = Path('data_extracted')
mcdh = Path('mcdh_extracted')
```

```python
# print all jpg files' filenames without full path in subdirectories of data
data_filenames = list(map(lambda x: x.name, list(data.rglob('*.jpg'))))
mcdh_filenames = list(map(lambda x: x.name, list(data.rglob('*.jpg'))))
```

```python
d_fpath = list(data.rglob('*.jpg'))
mcdh_fpath = list(mcdh.rglob('*.jpg'))
```

```python
d_fpath.sort()
mcdh_fpath.sort()
```

```python
data_filenames.sort()
mcdh_filenames.sort()
```

```python
print(len(data_filenames))
print(len(mcdh_filenames))
```

```python
print(data_filenames)
print(mcdh_filenames)
```

```python
# find all uncommon files in data and mcdh
uncommon_files = list(set(data_filenames) ^ set(mcdh_filenames))
uncommon_files.sort()
```

```python
# show the first 5 images in both files in 2 rows
from PIL import Image
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 5, figsize=(20, 10))

for i in range(5):
    image = Image.open(d_fpath[i])
    axs[0, i].imshow(image)
    axs[0, i].axis('off')
    axs[0, i].set_title(data_filenames[i])

    image = Image.open(mcdh_fpath[i])
    axs[1, i].imshow(image)
    axs[1, i].axis('off')
    axs[1, i].set_title(mcdh_filenames[i])

plt.show()
```
