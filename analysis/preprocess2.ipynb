```python
from pathlib import Path
import pandas as pd
```

```python
images_path = Path('images')
```

```python
for p in images_path.rglob('*.jpg'):
    # print name
    print(p.name)
    # print second element of the path
    print(p.parts[1])
```

```python
# construct data frame using filenames
# columns for:
#   path to file
#   second part of path is location
#   filename is a unix timestamp that is parsed as a datetime

records = []
for p in images_path.rglob('*.jpg'):
    records.append({
        'path': p,
        'location': p.parts[1],
        'timestamp': int(p.stem)
    })

df = pd.DataFrame.from_records(records)
```

```python
# convert timestamp to est datetime
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
```

```python
df['timestamp'] = df['timestamp'].dt.tz_convert('US/Eastern')
```

```python
# sort by timestamp
df = df.sort_values('timestamp')
```

```python
# range of timestamps
df['timestamp'].min(), df['timestamp'].max()
```
